## 1. Preprocessing Steps

### 1.1 Import Libraries & Download Resources

```python
import re
import json
import emoji
import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

### 1.2 Load Data

```python
with open('./dm-lab-2-private-competition/final_posts.json', 'r') as f:
    posts = json.load(f)

data = []
for item in posts:
    post_id = item['root']['_source']['post']['post_id']
    text = item['root']['_source']['post']['text']
    hashtags = item['root']['_source']['post']['hashtags']
    data.append({
        'id': post_id,
        'text': text,
        'hashtags': hashtags
    })

posts_df = pd.DataFrame(data)

data_id = pd.read_csv('./dm-lab-2-private-competition/data_identification.csv')
emotion = pd.read_csv('./dm-lab-2-private-competition/emotion.csv')

df = posts_df.merge(data_id, on='id')
df = df.merge(emotion, on='id', how='left')

train_df = df[df['split'] == 'train'].copy()
test_df = df[df['split'] == 'test'].copy()

print(f"Train: {len(train_df)}, Test: {len(test_df)}")
train_df.head()
```

### 1.3 Define Preprocessing Functions

The preprocessing pipeline consists of several key steps to clean and normalize the text data for emotion classification:

1. **Emoji to Text Conversion**: Emojis carry strong emotional signals (e.g., üòÄ indicates joy, üò¢ indicates sadness). Instead of removing them, we convert emojis to their text descriptions (e.g., üòÄ ‚Üí "grinning_face") to preserve this valuable emotional information.

2. **URL and Mention Removal**: URLs (http://...) and @mentions (@username) are noise that don't contribute to emotion detection, so we remove them from the text.

3. **Contraction Expansion**: Contractions like "won't", "can't", and "I'm" are expanded to their full forms ("will not", "cannot", "I am") to standardize the text and improve tokenization.

4. **Stopword Removal**: Common words like "the", "is", "at" appear frequently but carry little emotional meaning. However, we preserve negation words ("not", "no", "never") as they significantly impact sentiment (e.g., "not happy" vs "happy").

5. **Lemmatization**: Words are reduced to their base dictionary form (e.g., "running" ‚Üí "run", "happier" ‚Üí "happy"). This reduces vocabulary size and groups related words together, improving model generalization.

```python
stop_words = set(stopwords.words('english')) - {'not', 'no', 'never', 'very', 'so', 'too'}
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = emoji.demojize(text, delimiters=(" ", " "))
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#(\w+)', r'\1', text)
    contractions = {
        "won't": "will not", "can't": "cannot", "n't": " not",
        "'re": " are", "'ve": " have", "'ll": " will",
        "'d": " would", "'m": " am", "'s": " is"
    }
    for contraction, expansion in contractions.items():
        text = text.replace(contraction, expansion)
    text = re.sub(r'[^a-zA-Z\s_]', '', text)
    text = text.replace('_', ' ')
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

def preprocess_text(text):
    text = clean_text(text)
    tokens = word_tokenize(text)
    tokens = [w for w in tokens if w not in stop_words and len(w) > 2]
    tokens = [lemmatizer.lemmatize(w) for w in tokens]
    return ' '.join(tokens)
```

### 1.4 Apply Preprocessing

```python
train_df['clean_text'] = train_df['text'].apply(preprocess_text)
test_df['clean_text'] = test_df['text'].apply(preprocess_text)

print("raw text:")
print(train_df['text'].iloc[0])
print("\nclean version:")
print(train_df['clean_text'].iloc[0])
```

## 2. Feature Engineering Steps

Feature engineering extracts meaningful numerical representations from text data. We combine multiple feature types to capture different aspects of emotional expression:

1. **TF-IDF (Term Frequency-Inverse Document Frequency)**: Converts text into numerical vectors by measuring word importance. We use n-grams (1-3 words) to capture both individual words and phrases like "very happy" or "not good".

2. **Text Statistics**: Basic features such as text length, word count, and average word length can indicate emotional intensity (e.g., angry posts may be longer or use more punctuation).

3. **Punctuation Features**: Emotional text often contains repeated punctuation. Exclamation marks (!) may indicate joy or anger, question marks (?) may indicate surprise, and ellipsis (...) may indicate sadness.

4. **Emotion Keyword Matching**: We count occurrences of emotion-specific keywords (e.g., "happy", "love" for joy; "angry", "hate" for anger) to create direct emotion indicators.

5. **Sentiment Scores**: Using VADER sentiment analyzer to extract compound, positive, negative, and neutral scores, which provide additional emotional context.

### 2.1 Extract Text Statistics Features

```python
def extract_features(df):
    df['text_length'] = df['text'].str.len()
    df['word_count'] = df['text'].str.split().str.len()
    df['avg_word_length'] = df['text'].apply(
        lambda x: np.mean([len(w) for w in x.split()]) if len(x.split()) > 0 else 0
    )
    df['exclamation_count'] = df['text'].str.count('!')
    df['question_count'] = df['text'].str.count(r'\?')
    df['ellipsis_count'] = df['text'].str.count(r'\.\.\.')
    df['caps_ratio'] = df['text'].apply(
        lambda x: sum(1 for c in x if c.isupper()) / max(len(x), 1)
    )
    
    return df

train_df = extract_features(train_df)
test_df = extract_features(test_df)

print("Features extracted:")
print(train_df[['text_length', 'word_count', 'exclamation_count', 'caps_ratio']].head())
```

### 2.2 Emotion Keyword Matching (using NRC Emotion Lexicon)

I use the NRC Emotion Lexicon, which contains approximately 27,000 words annotated with 8 emotions (anger, fear, anticipation, trust, surprise, sadness, joy, disgust) and 2 sentiments (positive, negative).

```python
from nrclex import NRCLex

def add_nrc_emotions(df):
    emotions = ['anger', 'fear', 'surprise', 'sadness', 'joy', 'disgust']
    
    results = []
    for text in df['clean_text']:
        nrc = NRCLex(text)
        scores = nrc.raw_emotion_scores
        results.append({f'nrc_{emo}': scores.get(emo, 0) for emo in emotions})
    nrc_df = pd.DataFrame(results)
    df = pd.concat([df.reset_index(drop=True), nrc_df], axis=1)
    
    return df

train_df = add_nrc_emotions(train_df)
test_df = add_nrc_emotions(test_df)

print(train_df[['nrc_joy', 'nrc_anger', 'nrc_sadness', 'nrc_fear', 'nrc_surprise', 'nrc_disgust']].head(10))
print("\nNRC feature statistics:")
print(train_df[['nrc_joy', 'nrc_anger', 'nrc_sadness', 'nrc_fear', 'nrc_surprise', 'nrc_disgust']].sum())
```

### 2.3 TF-IDF Vectorization
Convert text to numerical vectors using TF-IDF with n-grams (1-3).

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1, 3),
    min_df=2,
    max_df=0.95,
    sublinear_tf=True
)

X_tfidf_train = tfidf.fit_transform(train_df['clean_text'])
X_tfidf_test = tfidf.transform(test_df['clean_text'])

print(f"TF-IDF shape - Train: {X_tfidf_train.shape}, Test: {X_tfidf_test.shape}")
```

### 2.4 Combine All Features
Merge TF-IDF features with extracted numerical features.

```python
from scipy.sparse import hstack

numerical_features = [
    'text_length', 'word_count', 'avg_word_length',
    'exclamation_count', 'question_count', 'ellipsis_count', 'caps_ratio',
    'nrc_joy', 'nrc_anger', 'nrc_sadness', 
    'nrc_fear', 'nrc_surprise', 'nrc_disgust'
]

X_train = hstack([X_tfidf_train, train_df[numerical_features].values])
X_test = hstack([X_tfidf_test, test_df[numerical_features].values])

y_train = train_df['emotion']

print(f"Final feature shape - Train: {X_train.shape}, Test: {X_test.shape}")
```

## 3. Model Implementation Steps

## 3. Model Implementation Steps

I use an ensemble approach combining multiple classifiers to improve prediction accuracy. The models include:

1. **Logistic Regression**: A fast, interpretable linear model that works well with high-dimensional TF-IDF features. We use L2 regularization and class weights to handle imbalanced data.

2. **Multinomial Naive Bayes**: A probabilistic classifier that performs well on text classification tasks. It's particularly effective with TF-IDF features.

3. **Random Forest**: An ensemble of decision trees that can capture non-linear patterns and feature interactions.

4. **XGBoost**: A gradient boosting algorithm known for strong performance in classification tasks. It handles class imbalance well with the `scale_pos_weight` parameter.

Finally, I combine predictions using **soft voting** (averaging predicted probabilities) to leverage the strengths of each model.

### 3.1 Train Individual Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')

le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)

print("1. Logistic Regression")
lr_model = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
lr_model.fit(X_train, y_train)
lr_scores = cross_val_score(lr_model, X_train, y_train, cv=3, scoring='f1_macro')
print(f"   CV F1: {lr_scores.mean():.4f}")

print("2. Naive Bayes")
nb_model = MultinomialNB(alpha=0.1)
nb_model.fit(X_train, y_train)
nb_scores = cross_val_score(nb_model, X_train, y_train, cv=3, scoring='f1_macro')
print(f"   CV F1: {nb_scores.mean():.4f}")

print("3. Random Forest")
rf_model = RandomForestClassifier(n_estimators=100, max_depth=15, class_weight='balanced', random_state=42, n_jobs=-1)
rf_model.fit(X_train, y_train)
rf_scores = cross_val_score(rf_model, X_train, y_train, cv=3, scoring='f1_macro')
print(f"   CV F1: {rf_scores.mean():.4f}")

print("4. XGBoost (training only, no CV)")
xgb_model = XGBClassifier(n_estimators=100, max_depth=5, learning_rate=0.1, random_state=42, n_jobs=-1)
xgb_model.fit(X_train, y_train_encoded)
```

```python
lr_proba = lr_model.predict_proba(X_test)
nb_proba = nb_model.predict_proba(X_test)
rf_proba = rf_model.predict_proba(X_test)
xgb_proba = xgb_model.predict_proba(X_test)

print("LR classes:", lr_model.classes_)
print("XGB classes:", le.classes_)

weights = [0.4, 0.2, 0.2, 0.2]
ensemble_proba = (
    weights[0] * lr_proba + 
    weights[1] * nb_proba + 
    weights[2] * rf_proba + 
    weights[3] * xgb_proba
)

ensemble_pred_idx = np.argmax(ensemble_proba, axis=1)
ensemble_pred = lr_model.classes_[ensemble_pred_idx]

print("\nPrediction distribution:")
print(pd.Series(ensemble_pred).value_counts())
```

```python
submission = pd.DataFrame({
    'id': test_df['id'],
    'emotion': ensemble_pred
})

submission.to_csv('submission.csv', index=False)

print("Submission saved!")
print(submission.head(10))
print(f"\nTotal predictions: {len(submission)}")
```
